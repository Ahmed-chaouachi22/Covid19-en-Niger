# Description of the Different Types of Variables

The dataset contains several categories of variables describing the individuals, their health condition, and their behaviors related to disease prevention.

---

## i). Demographic and Geographic Variables

Variables :

`Code, Longitude, Latitude, Quartier_corrige, Population`

**Description**

These variables describe the **geographic location and spatial distribution of individuals** in the study.

* **Code** : Unique identifier assigned to each individual in the dataset.
* **Longitude / Latitude** : Geographic coordinates indicating the precise location of the individual or household.
* **Quartier_corrige** : Name of the neighborhood where the individual lives. This variable is used to analyze the **spatial distribution of cases**.
* **Population** : Estimated population of the neighborhood.

These variables are important for **geographic analysis and identifying spatial clusters of cases**.

---

## 2i). Socio-Demographic Characteristics

Variables :

`Sexe, Age, Categorie_age, Profession, Statut_matrimonial`

**Description**

These variables describe the **social and demographic characteristics of the individuals** included in the study population.

* **Sexe** : Gender of the individual (male or female).
* **Age** : Age of the individual in years.
* **Categorie_age** : Age group classification used to facilitate population analysis.
* **Profession** : Occupation of the individual. This variable can reflect differences in exposure risk.
* **Statut_matrimonial** : Marital status of the individual (single, married, etc.).

These variables help describe the **structure of the study population and identify demographic patterns related to infection risk**.

---

## 3i). Comorbidities

Variables :

`Diabte, HTA, Maladie_cardiaque`

**Description**

These variables represent **pre-existing health conditions** that may influence susceptibility to infection or severity of the disease.

* **Diabte** : Indicates whether the individual has diabetes.
* **HTA** : Indicates whether the individual has hypertension.
* **Maladie_cardiaque** : Indicates the presence of heart disease.

Comorbidities are important because they may **increase vulnerability to infectious diseases and complications**.

---

## 4i). Symptoms

Variables :

Céphalée, Douleurs articulaires, Fièvre, Douleurs musculaires,
Douleurs abdominales, Toux, Rhinorrhée, Anosmie, Ageusie,
Frissons, Nausées, Dyspnée, Vomissements, Sueurs,
Eruption cutanée, Odynophagie, Conjonctivite, Rhinorragie

**Description**

These variables indicate whether individuals reported **clinical symptoms associated with infection**.

Examples include :

* **Fièvre (Fever)**
* **Toux (Cough)**
* **Dyspnée (Difficulty breathing)**
* **Anosmie (Loss of smell)**
* **Ageusie (Loss of taste)**

These variables help identify **symptom patterns associated with serological positivity**.

---

## 5i). Preventive Behaviors

Variables :

Port du masque à domicile, Port masque exterieur,
Usage du Gel hydroalcoolique, Usage savon et eau

**Description**

These variables describe **preventive health behaviors adopted by individuals to reduce the risk of infection**.

* **Port du masque à domicile** : Wearing a mask inside the home.
* **Port masque exterieur** : Wearing a mask in public spaces.
* **Usage du Gel hydroalcoolique** : Use of hand sanitizer.
* **Usage savon et eau** : Handwashing with soap and water.

## 4- Destribution du Serologie

```python
def plot_serologie_by_variable(df, variable):
    """
    Crée un graphique à barres horizontales comparant le total d'individus
    et les cas positifs pour chaque modalité d'une variable catégorielle.

    Paramètres :
    - df : DataFrame contenant les données.
    - variable : nom de la colonne catégorielle à analyser.
    - show_total_pct : booléen, si True ajoute le pourcentage du total de chaque
                       modalité par rapport à l'ensemble (par défaut False).
    """
    show_total_pct=True
    # Agrégation des données
    stats = df.groupby(variable).agg(
        Total=('Serologie', 'count'),
        Positifs=('Serologie', lambda x: (x == 'Positif').sum())
    ).reset_index()

    stats['Taux_positivite'] = (stats['Positifs'] / stats['Total'] * 100).round(1)

    if show_total_pct:
        total_global = len(df)
        stats['Pct_total'] = (stats['Total'] / total_global * 100).round(1)

    # Tri par nombre de positifs décroissant (optionnel)
    stats = stats.sort_values('Positifs', ascending=False)

    # Création des labels
    stats['label_total'] = stats['Total'].astype(str)
    if show_total_pct:
        stats['label_total'] += ' (' + stats['Pct_total'].astype(str) + '%)'

    stats['label_positifs'] = (stats['Positifs'].astype(str) +
                               ' (' + stats['Taux_positivite'].astype(str) + '%)')

    # Construction du graphique
    fig = make_subplots(rows=1, cols=1)

    # Barres du total
    fig.add_trace(go.Bar(
        y=stats[variable],
        x=stats['Total'],
        name='Total',
        orientation='h',
        marker_color="#4e98bc",
        text=stats['label_total'],
        textposition='auto',
        hovertemplate='<b>%{y}</b><br>Total: %{x}<extra></extra>'
    ))

    # Barres des positifs
    fig.add_trace(go.Bar(
        y=stats[variable],
        x=stats['Positifs'],
        name='Positifs',
        orientation='h',
        marker_color="#e73c3c",
        text=stats['label_positifs'],
        textposition='auto',
        hovertemplate='<b>%{y}</b><br>Positifs: %{x}<br>Taux: %{customdata}%<extra></extra>',
        customdata=stats['Taux_positivite']
    ))

    # Mise en page
    fig.update_layout(
        title=f'Individus et cas positifs par {variable}',
        xaxis_title='Effectif',
        yaxis_title=variable,
        barmode='stack',
        template='plotly_white',
        height=600
    )

    fig.show()
    

```

## 5. Analyse des symptômes

```python
print("\n=== ANALYSE DES SYMPTÔMES ===\n")

n_total = len(df)
# Vérifier que toutes les colonnes existent
symptom_cols = [col for col in symptomes if col in df.columns]
print(f"Nombre de symptômes analysés : {len(symptom_cols)}")

# Fonction pour encoder les symptômes : Oui/Rare/Fréquent -> 1, Non -> 0
def encode_symptom(val):
    if pd.isna(val):
        return np.nan
    if val in ['Oui', 'Rare', 'Fréquent']:
        return 1
    else:
        return 0

# Créer une matrice binaire des symptômes
symptom_matrix = df[symptom_cols].applymap(encode_symptom)

# Fréquence de chaque symptôme (nombre de personnes avec présence)
symptom_freq = symptom_matrix.sum().sort_values(ascending=False).reset_index()
symptom_freq.columns = ['Symptôme', 'Effectif']
symptom_freq['Pourcentage'] = (symptom_freq['Effectif'] / n_total * 100).round(1)

print("\nFréquence des symptômes :")
print(symptom_freq.to_string(index=False))

```

```python
# Garder seulement les lignes avec sérologie positive
df_pos = df[df['Serologie'] == 'Positif'].copy()
n_pos = len(df_pos)
print(f"Nombre de cas positifs : {n_pos}")
print(f"Soit {n_pos/len(df)*100:.1f}% de l'échantillon total")

# =============================================================================
# 2. Définition des catégories de variables
# =============================================================================
demographic_vars = ['Sexe', 'Age', 'Categorie_age', 'Profession', 'Statut_matrimonial']
comorbid_vars = ['Diabte', 'HTA', 'Maladie_cardiaque']
symptom_vars = [
    'Céphalée', 'Douleurs articulaires', 'Fièvre', 'Douleurs musculaires',
    'Douleurs abdominales', 'Toux', 'Rhinorrhée', 'Anosmie', 'Ageusie',
    'Frissons', 'Nausées', 'Dyspnée', 'Vomissements', 'Sueurs',
    'Eruption cutanée', 'Odynophagie', 'Conjonctivite', 'Rhinorragie'
]
preventive_vars = [
    'Port du masque à domicile', 'Port masque exterieur',
    'Usage du Gel hydroalcoolique', 'Usage savon et eau'
]
geo_vars = ['Quartier_corrige', 'Population']

# Ne garder que les colonnes existantes
symptom_vars = [col for col in symptom_vars if col in df.columns]
comorbid_vars = [col for col in comorbid_vars if col in df.columns]
preventive_vars = [col for col in preventive_vars if col in df.columns]

# Fonction d'encodage binaire pour les symptômes/comorbidités
def encode_binary(val):
    return 1 if val in ['Oui', 'Rare', 'Fréquent'] else 0

# =============================================================================
# 3. Profil démographique des positifs
# =============================================================================
print("\n" + "="*60)
print("PROFIL DÉMOGRAPHIQUE DES CAS POSITIFS")
print("="*60)

# Sexe
sexe_counts = df_pos['Sexe'].value_counts()
sexe_percent = df_pos['Sexe'].value_counts(normalize=True) * 100
print("\nSexe :")
print(sexe_counts)
print("Pourcentage :")
print(sexe_percent.round(1))

fig_sexe = px.pie(
    names=sexe_counts.index, values=sexe_counts.values,
    title=f"Répartition par sexe parmi les positifs (n={n_pos})",
    color_discrete_sequence=['#3498db', '#e91e63'],
    hole=0.3
)
fig_sexe.update_traces(textposition='inside', textinfo='percent+label')
fig_sexe.show()

# Âge
print(f"\nÂge : moyenne = {df_pos['Age'].mean():.1f}, médiane = {df_pos['Age'].median():.1f}")
fig_age = px.histogram(
    df_pos, x='Age', nbins=30, marginal='box',
    title=f"Distribution de l'âge des positifs (n={n_pos})",
    color_discrete_sequence=['#e74c3c']
)
fig_age.add_vline(x=df_pos['Age'].mean(), line_dash="dash", line_color="black")
fig_age.show()

# Catégorie d'âge
order_age = ['Enfant', 'Adolescent', 'Adulte', 'Personne âgée']
agecat_counts = df_pos['Categorie_age'].value_counts().reindex(order_age)
print("\nCatégorie d'âge :")
print(agecat_counts)

fig_agecat = px.bar(
    x=agecat_counts.index, y=agecat_counts.values,
    title=f"Répartition par catégorie d'âge (positifs)",
    labels={'x': '', 'y': 'Effectif'},
    color=agecat_counts.index,
    color_discrete_sequence=px.colors.qualitative.Pastel,
    text=agecat_counts.values
)
fig_agecat.update_traces(textposition='auto')
fig_agecat.show()

# Profession
prof_counts = df_pos['Profession'].value_counts()
print("\nProfession :")
print(prof_counts)

fig_prof = px.bar(
    x=prof_counts.values, y=prof_counts.index, orientation='h',
    title=f"Répartition par profession (positifs)",
    labels={'x': 'Effectif', 'y': ''},
    color=prof_counts.index,
    color_discrete_sequence=px.colors.qualitative.Set2,
    text=prof_counts.values
)
fig_prof.update_traces(textposition='auto')
fig_prof.update_layout(yaxis={'categoryorder': 'total ascending'})
fig_prof.show()

# Statut matrimonial
statut_counts = df_pos['Statut_matrimonial'].value_counts()
print("\nStatut matrimonial :")
print(statut_counts)

fig_statut = px.bar(
    x=statut_counts.index, y=statut_counts.values,
    title=f"Répartition par statut matrimonial (positifs)",
    labels={'x': '', 'y': 'Effectif'},
    color=statut_counts.index,
    color_discrete_sequence=px.colors.qualitative.Set3,
    text=statut_counts.values
)
fig_statut.update_traces(textposition='outside')
fig_statut.show()

# =============================================================================
# 4. Comorbidités chez les positifs
# =============================================================================
print("\n" + "="*60)
print("COMORBIDITÉS CHEZ LES POSITIFS")
print("="*60)

# Encodage binaire des comorbidités
comor_matrix = df_pos[comorbid_vars].applymap(encode_binary)
comor_freq = comor_matrix.sum().sort_values(ascending=False)
comor_percent = (comor_freq / n_pos * 100).round(1)

comor_df = pd.DataFrame({
    'Comorbidité': comor_freq.index,
    'Effectif': comor_freq.values,
    'Pourcentage': comor_percent.values
})
print("\nPrévalence des comorbidités parmi les positifs :")
print(comor_df.to_string(index=False))

fig_comor = px.bar(
    comor_df,
    x='Comorbidité',
    y='Effectif',
    text='Effectif',
    title=f"Comorbidités chez les positifs (n={n_pos})",
    color='Comorbidité',
    color_discrete_sequence=px.colors.qualitative.Set1,
    hover_data={'Pourcentage': True}
)
fig_comor.update_traces(
    texttemplate='%{text} (%{customdata:.1f}%)',
    textposition='auto',
    customdata=comor_df['Pourcentage'].values
)
fig_comor.show()

# =============================================================================
# 5. Symptômes chez les positifs
# =============================================================================
print("\n" + "="*60)
print("SYMPTÔMES CHEZ LES POSITIFS")
print("="*60)

symptom_matrix = df_pos[symptom_vars].applymap(encode_binary)
symptom_freq = symptom_matrix.sum().sort_values(ascending=False)
symptom_percent = (symptom_freq / n_pos * 100).round(1)

symptom_df = pd.DataFrame({
    'Symptôme': symptom_freq.index,
    'Effectif': symptom_freq.values,
    'Pourcentage': symptom_percent.values
})
print("\nFréquence des symptômes parmi les positifs :")
print(symptom_df.to_string(index=False))

fig_symp = px.bar(
    symptom_df,
    x='Effectif',
    y='Symptôme',
    orientation='h',
    text='Effectif',
    title=f"Symptômes les plus fréquents chez les positifs (n={n_pos})",
    color='Effectif',
    color_continuous_scale='Reds',
    hover_data={'Pourcentage': True}
)
fig_symp.update_traces(
    texttemplate='%{text} (%{customdata:.1f}%)',
    textposition='outside',
    customdata=symptom_df['Pourcentage'].values
)
fig_symp.update_layout(yaxis={'categoryorder': 'total ascending'})
fig_symp.show()

# Nombre de symptômes par individu positif
symptom_count = symptom_matrix.sum(axis=1)
print(f"\nNombre de symptômes par individu positif : moyenne = {symptom_count.mean():.1f}, médiane = {symptom_count.median():.0f}")
fig_count = px.histogram(
    x=symptom_count, nbins=15,
    title="Distribution du nombre de symptômes chez les positifs",
    labels={'x': 'Nombre de symptômes', 'y': 'Effectif'},
    color_discrete_sequence=['#e74c3c']
)
fig_count.add_vline(x=symptom_count.mean(), line_dash="dash", line_color="black")
fig_count.show()

# =============================================================================
# 6. Comportements préventifs chez les positifs
# =============================================================================
print("\n" + "="*60)
print("COMPORTEMENTS PRÉVENTIFS CHEZ LES POSITIFS")
print("="*60)

preventive_list = []
for col in preventive_vars:
    tmp = df_pos[col].value_counts().reset_index()
    tmp.columns = ['Réponse', 'Effectif']
    tmp['Comportement'] = col
    tmp['Pourcentage'] = (tmp['Effectif'] / n_pos * 100).round(1)
    preventive_list.append(tmp)
preventive_df = pd.concat(preventive_list, ignore_index=True)

print("\nDistribution des comportements préventifs parmi les positifs :")
print(preventive_df.to_string(index=False))

fig_prev = px.bar(
    preventive_df,
    x='Comportement',
    y='Effectif',
    color='Réponse',
    barmode='group',
    title=f"Comportements préventifs chez les positifs (n={n_pos})",
    text='Effectif',
    color_discrete_sequence=px.colors.qualitative.Set2,
    hover_data={'Pourcentage': True}
)

fig_prev.show()

# =============================================================================
# 7. Distribution géographique des positifs
# =============================================================================
print("\n" + "="*60)
print("DISTRIBUTION GÉOGRAPHIQUE DES POSITIFS")
print("="*60)

# Agrégation par quartier
quartier_stats = df.groupby('Quartier_corrige').agg(
    Total=('Serologie', 'count'),
    Positifs=('Serologie', lambda x: (x == 'Positif').sum())
).reset_index()
quartier_stats['Taux_positivite'] = (quartier_stats['Positifs'] / quartier_stats['Total'] * 100).round(1)
quartier_stats = quartier_stats.sort_values('Positifs', ascending=False)

# Top 10 quartiers avec le plus de positifs
top10_pos = quartier_stats.head(10)
print("\nTop 10 des quartiers avec le plus de cas positifs :")
print(top10_pos[['Quartier_corrige', 'Positifs', 'Total', 'Taux_positivite']].to_string(index=False))

fig_geo = px.bar(
    top10_pos,
    x='Positifs',
    y='Quartier_corrige',
    orientation='h',
    text='Positifs',
    title="Top 10 des quartiers avec le plus de cas positifs",
    labels={'Positifs': "Nombre de cas positifs", 'Quartier_corrige': ''},
    color='Taux_positivite',
    color_continuous_scale='Reds',
    hover_data={'Total': True, 'Taux_positivite': True}
)
fig_geo.update_traces(textposition='auto')
fig_geo.update_layout(yaxis={'categoryorder': 'total ascending'})
fig_geo.show()

# Carte des cas positifs (échantillon)
# On prend un échantillon pour ne pas surcharger
sample_pos = df_pos.sample(min(1000, len(df_pos)), random_state=42)
fig_map = px.scatter_mapbox(
    sample_pos,
    lat='Latitude',
    lon='Longitude',
    color_discrete_sequence=['#e74c3c'],
    zoom=11,
    mapbox_style='carto-positron',
    title=f"Localisation des cas positifs (échantillon de {len(sample_pos)})",
    hover_data={'Quartier_corrige': True, 'Age': True, 'Sexe': True}
)
fig_map.update_layout(margin=dict(l=0, r=0, t=40, b=0))
fig_map.show()

# =============================================================================
# 8. Synthèse : profil type du cas positif
# =============================================================================
print("\n" + "="*60)
print("SYNTHÈSE : PROFIL TYPE DU CAS POSITIF")
print("="*60)

print(f"\n- Âge moyen : {df_pos['Age'].mean():.1f} ans")
print(f"- Sexe : {sexe_percent['Femme']:.1f}% de femmes, {sexe_percent['Homme']:.1f}% d'hommes")
print(f"- Catégorie d'âge la plus représentée : {agecat_counts.idxmax()} ({agecat_counts.max()} personnes)")
print(f"- Profession la plus fréquente : {prof_counts.idxmax()} ({prof_counts.max()} personnes)")
print(f"- Statut matrimonial le plus fréquent : {statut_counts.idxmax()} ({statut_counts.max()} personnes)")

print("\nComorbidités les plus fréquentes :")
for _, row in comor_df.head(3).iterrows():
    print(f"  - {row['Comorbidité']} : {row['Effectif']} ({row['Pourcentage']}%)")

print("\nSymptômes les plus fréquents :")
for _, row in symptom_df.head(5).iterrows():
    print(f"  - {row['Symptôme']} : {row['Effectif']} ({row['Pourcentage']}%)")

print("\nComportements préventifs :")
for col in preventive_vars:
    modalite = df_pos[col].mode()[0]  # modalité la plus fréquente
    count = (df_pos[col] == modalite).sum()
    print(f"  - {col} : le plus souvent '{modalite}' ({count} personnes, {count/n_pos*100:.1f}%)")

print("\nQuartiers avec le plus de cas :")
for _, row in top10_pos.head(3).iterrows():
    print(f"  - {row['Quartier_corrige']} : {row['Positifs']} cas (taux {row['Taux_positivite']}%)")

print("\n" + "="*60)
print("ANALYSE DES CAS POSITIFS TERMINÉE")
print("="*60)
```
